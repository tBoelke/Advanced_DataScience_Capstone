# 1. This is a first exploration of data #
Ok, let's import some libraries...

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

## 1.1 Now let's load the data ##
I will use the well known Iris dataset from UCI

```python
df = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data', header = None)
```

## 1.2 First look at the data ##
For a very first look I use the .head() and .info() methods

```python
df.head(5)
```

```python
df.info()
```

**As you can see:**
* the dataset has 150 rows and 5 columns and no header
* there are no missing values
* we have 4 numerical features
* we have 1 label


## 1.3 First look at some basic statistics using 3D Scatterplots and .describe() method ##

**In order to get a very first impression lets plot some 3D scatterplots.**

```python
plot3D = plt.figure().gca(projection='3d')
plot3D.scatter(df[[0]],df[[1]], df[[2]])
plot3D.set_xlabel('Feature 1')
plot3D.set_ylabel('Feature 2')
plot3D.set_zlabel('Feature 3')
plt.show()
```

```python
plot3D = plt.figure().gca(projection='3d')
plot3D.scatter(df[[0]],df[[1]], df[[3]])
plot3D.set_xlabel('Feature 1')
plot3D.set_ylabel('Feature 2')
plot3D.set_zlabel('Feature 4')
plt.show()
```

```python
plot3D = plt.figure().gca(projection='3d')
plot3D.scatter(df[[1]],df[[2]], df[[3]])
plot3D.set_xlabel('Feature 2')
plot3D.set_ylabel('Feature 3')
plot3D.set_zlabel('Feature 4')
plt.show()
```
